
-   Explain why your solution works


```python
# Your answer here
# The main function creates a list to store the paths and calls the helper function, passing the root, an empty list for the path, and the list of paths
# The helper function first checks for the base case (the node does not exist)
#   then it appends the current node value to the path list
#   finally, it checks to see if there are no children. in this case it will append the list of paths to the solution list
#   and then it will recursively call the left and right children
# This allows the code to traverse from root to leaf on each path and store the values of each node on the path
```


-   Explain the problem’s time and space complexity
