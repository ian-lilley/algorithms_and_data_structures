
-   Copy the solution your partner wrote. 


```python
# Your answer here
from typing import List
def missing_num(nums: List[int]) -> List[int]:
        num_set = set(nums)
        n = len(nums)
        missing = []
        for i in range(n + 1):
            if i not in num_set:
                missing.append(i)
        return missing if missing else [-1]
```


-   Explain why their solution works in your own words.


-   Explain the problem’s time and space complexity in your own words.


```python
# Your answer here
# The code must run through every item in the list when converting to a set
#   and also in the for loop, so the time complexity is O(n)
# There are potentially three copies of the list needed: 
#   the original list, the list converted to a set, and the list of missing values
#   worst case scenario is n for each list or 3n, so the space complexity is O(n)
```
